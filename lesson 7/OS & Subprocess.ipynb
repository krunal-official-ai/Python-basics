# ✅ map( ) – Transform Elements
# Transforms each element of an iterable.

# map() applies the SAME function to EACH item in a collection

# Think of it like:

# “Take every value → apply logic → give me new values”

```python
# map(function, iterable)

# function → what to do

# iterable → data (list, tuple, etc.)

# map() returns an iterator, so we usually convert it to list().
```

map(function, iterable)

```python
## WITHOUT MAP 
nums = [1, 2, 3, 4]
result = []

for n in nums:
    result.append(n * 2)

print(result)
```

```python
## WITH map 
nums = [1, 2, 3, 4]
result = list(map(lambda x: x * 2, nums))
print(result)
```

```python
# map(function , argument)

def hello(n):
    return n ** 2

map(hello,[1,2,3,4])
```

```python
def hello(n):
    return n ** 2

for i  in map(hello,[1,2,3,4]):
    print(i)
```

```python
map_data = list(map(hello,[1,2,3,4]))

map_data 
```

```python
# map functions with lambda functions

map_data = list(map(lambda x : x**2 , [1,2,3,4]))

map_data
```

```python
map_data = list(map(lambda x : x**3 , [1,2,3,4]))

map_data
```

# ✅ filter( ) – Filter Based on Condition
# Selects elements based on condition

```python
filter(function, iterable)
```

```python
marks = [35, 70, 90, 20, 55]

passed = list(filter(lambda x: x >= 40, marks))

print(passed)
```

# ✅ reduce( ) – Reduce to a Single Value
# Reduces iterable into single value
# reduce() takes many values and reduces them to ONE single value by repeatedly applying a function.
# Many → One

```python
from functools import reduce
reduce(function, iterable)
```

```python
nums = [1, 2, 3, 4]
```

```python
## WITHOUT REDUCE 
total = 0
for n in nums:
    total += n
print(total)
```

```python
from functools import reduce

total = reduce(lambda a, b: a + b, nums)
print(total)
```

```python
nums = [1, 2, 3, 4]

Step 1: a=1, b=2 → result = 3
Step 2: a=3, b=3 → result = 6
Step 3: a=6, b=4 → result = 10
```
