# %A  - Formate specifier

```python
O = datetime.datetime.now()
print(O.strftime("%b"))
```

```python
#| Code | Meaning                        | Example  |
#| ---- | ------------------------------ | -------- |
#|  %d  | Day of the month (01–31)       |  28      |
#|  %a  | Abbreviated weekday name       |  Mon     |
#|  %A  | Full weekday name              |  Monday  |
#|  %w  | Weekday (0=Sunday, 6=Saturday) |  1       |
#|  %j  | Day of the year (001–366)      |  209     |
#|  %m  | Month (01–12)                  |  07      |
#|  %b  | Abbreviated month name         |  Jul     |
#|  %B  | Full month name                |  July    |
#|  %y  | Year without century (00–99)   |  25      |
#|  %Y  | Year with century              |  2025    |

```

# CODE - Parsing Strings to Dates
# Converting string → datetime

```python
from datetime import datetime

date_str = "2025-07-28 14:30:00"

date_obj = datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S")

print(date_obj)
```

# CODE - Date Arithmetic showcase of tomorrow's data

```python
from datetime import timedelta

today = datetime.now()

tomorrow = today + timedelta(days=1)


print("Today    :", today)

print("Tomorrow :", tomorrow)
```

# CODE - Calculate Age or Duration

```python
birth_date = datetime(1993,3,3)

today = datetime.now()

age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))

print("Age:", age)
```

# CODE - time.ctime([secs])



```python
print(time.ctime())          # Current time
print(" ")
print(time.ctime(1))         # Epoch time
```

**sec: This is an optional argument representing the number of seconds since the epoch (January 1, 1970, 00:00:00 UTC). If omitted or None, time.ctime() uses the current local time as returned by time.time().**


**Return Value: It returns a string in the format: Day Mon DD HH:MM:SS YYYY (e.g., "Wed Oct 1 10:07:00 2025").**
