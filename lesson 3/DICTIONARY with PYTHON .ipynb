# A dictionary (type dict) in Python is a collection of key: value pairs.
# Keys are unique and immutable (like strings, numbers, tuples), while values can be any data type—even other dictionaries or lists

```python
# DICTIONARY :

X = { 'name'  :'Robort downy Jr.',
      'age'   : '34',
      'number':'67549387',
      'gender':'m'
     }

```

```python
print(X)
print(type(X))
```

```python
print("Key : ",X.keys())
```

```python
print("Values : ",X.values())
```

```python
print(X['name'])
```

```python
X['name'] = 'Ironman' # update method
print("Update",X)
```

```python
X['rollnumber'] = 909090909 # Add method  - To add new Element in dictionary
print("Insert/append",X)
```

```python
del X['age']
print ("del : ",X)
```

```python
print(X.get('name')) #  Get method to check keys - element is inside our dictionary.
print(X.get('age'))
print(X.get('age',"AGE IS NOT INSIDE"))
print(X.get('enrollment',"Not in Dictionary"))
```
