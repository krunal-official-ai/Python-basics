## 1. What is a List of Dictionaries?
A list of dictionaries is a common Python structure used to store tabular data (rows & columns).

```python
students = [
    {"id": 1, "name": "Amit", "marks": 85},
    {"id": 2, "name": "Kiran", "marks": 90},
    {"id": 3, "name": "Riya", "marks": 78}
]

students
```

## 2. Reading CSV as List of Dictionaries

```python
import csv

students = []
with open(r"C:\Users\krunal sir\Downloads\students.csv",'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        students.append(row)

students
```

## 3. Accessing Data

```python
# Print all names
for s in students:
    print(s["name"])
```

## 4. Filtering Data

```python
# Students scoring above 80
top_students = [s for s in students if int(s["marks"]) > 80]
top_students
```

## 5. Writing Back to CSV

```python
import csv

with open("C:\\Users\\krunal sir\\Downloads\\students.csv", "w", newline="") as file:
    fieldnames = ["id", "name", "marks", "city"]
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(students)
```

# 6.Count the Data

```python
grade_count = {}

for student in students:
    grade = student["marks"]

    if grade in grade_count:
        grade_count[grade] += 1
    else:
        grade_count[grade] = 1

print(grade_count)
```
