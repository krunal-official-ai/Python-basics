# ✅ Testing Strings

```python
#isalnum() method
print("isalnum() method")
# returns True if all the characters in the string is alphanumeric
# (a string which contains either number or alphabets or both). Otherwise False.

s = "Abiteof python"
print(s.isalnum())
print("123".isalnum())
print("abc".isalnum())
print("abc123".isalnum())
```

```python
#isalpha() method
print("isalpha() method")
# returns True if all the characters in the string are alphabets.
# Otherwise False.
print("123".isalpha())
print("zyz".isalpha()
print("$$$$".isalpha())
print("abc233".isalpha())
```

```python
#isdigit() method
print("isdigit() method")
# returns True if all the characters in the string are digits.
# Otherwise False.
print("name101".isdigit())
print("101".isdigit())
print("101 ".isdigit())
print("101.129".isdigit())
```

```python
# islower() and isupper() method
print("islower()")
# returns True if all the characters in the string are in lowercase.
# Otherwise False.
print("abc".islower())
```

```python
print("isupper()")
# returns True if all the characters in the string are in uppercase.
# Otherwise False.
print("ABC".isupper())
```

# ✅ Searching and Replacing Strings


```python
# Searching and Replacing Strings

##  endswith - Returns True if string ends with substring sub. Otherwise False.
s = "abcdef"
s.endswith("ef")
```

```python
##  Startswith  - Returns True if string starts with substring sub. Otherwise False.

print ("python".startswith("py"))
```

```python
##  find - Returns the lowest index of the string where substring sub is found.
##          If substring sub is not found -1 is returned.
print("Learning with Python".find("i"))
```

```python
##  rfind - Returns the highest index of the string where substring sub is found.
##          If substring sub is not found -1 is returned.
print("Learning whit Python".rfind("i"))
```

```python
## find - It returns the number of occurrences of substring sub found in the string.
##.       If no occurrences found  -1  is returned.
print("Learning Python".find("at"))
```

```python
## count - It returns a new string after replacing old substring with new.
##         Notice that it does not change the object on which it is called.
print("procrastination is the thief of time".count("ti"))
```
