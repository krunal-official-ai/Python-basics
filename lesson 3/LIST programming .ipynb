# APPEND | INSERT | DELETE | UPDATE | INDEX | CLEAR | SORT | REVERSE | POP

```python
# method Append
thislist = ["apple", "banana", "cherry"]

thislist.append("orange") #method Append

print("method Append")

print(thislist)
```

```python
# method insert
thislist = ['apple', 'mango', 'banana', 'cherry', 'orange']

thislist.insert(1, "Pinaple") #method insert

print("method insert")

print(thislist)
```

```python
# method Delete
thislist = ['apple', 'Pinaple', 'cherry', 'banana' ,'orange']

thislist.remove("banana") # method Delete

print("method Delete")

print(thislist)
```

```python
# method update
thislist = ['apple', 'Robotics', 'cherry', 'orange']

thislist[1] = 'lemon' # method update

print("method update")

print(thislist)
```

```python
# method index
list1 = ["A", "B", "C", "D", "E", "F", "G" ,"H", "I"]

print("method index ")

print(list1.index("B"))

print(list1.index("D"))

print(list1.index("F"))

print(list1.index("I"))
```

```python
# method clear
list1 = [1, 2, 3, 4, 5, 6, 1, 2, 3]

print("method clear")

list1.clear() # method clear

print(list1)
```

```python
# sort the list

list1 = [12, -2, 3, 4, 100, 50]

list1.sort() # sort the list in ascending order\n",

print("method sort")

print(list1)

list1.reverse() # sort the list in descending order\n",

print("method reverse")

print(list1)
```

```python
# remove the element at index 4
list1 = [12, -2, 3, 4, 100, 50]

list1.pop(3) # remove the Index element

print("method pop")
print(list1)
```
