# ## SET with python  
# A set is a built-in Python collection type.

# It is an unordered, mutable data structure that stores unique elements (no duplicates).

# Defined using curly braces { } or the set( ) constructor.

# Sets just like a list ?  However, they are different  from list in the following ways:

# 1. Each element inside a set must be unique.
# 2. Elements in a set are stored in NO particular order.


```python
# Empty set
s = set()
```

# Note. { } would create a dict, not a set

```python
s1 = {1, 2, 3}
s2 = set([1, 2, 3, 3])

print(s1)
print(s2)
```

# #Union (A ∪ B)

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(A | B)   # Union → {1,2,3,4,5,6}
```

# #Intersection (A ∩ B)

```python
print(A & B)
```

# #Difference (A − B)

```python
print(A - B)
```

# #Difference (B − A)

```python
print(B - A)
```

# #Symmetric Difference (A △ B)

```python
print(A ^ B)
```

# Add / Update

```python
s = {1, 2, 3}

s.add(4)            # {1,2,3,4}
s.update([5, 6])    # {1,2,3,4,5,6}
```

```python
print(s)
```

# Remove elements

```python
s = {1, 2, 3, 4, 5, 6}


s.remove(2)         # Removes 2 (Error if not found)
print(s)

s.discard(10)       # Safe remove (no error if missing)
print(s)

s.pop()             # Removes random element
print(s)

s.clear()           # Empty the set
print(s)
```

```python
nums = {1, 2, 3, 4, 5}

print(len(nums))   # 5

print(max(nums))   # 5

print(min(nums))   # 1

print(sum(nums))   # 15
```

# frozenset

```python
fst = frozenset([1, 2, 3, 2])
print(fst)  # frozenset({1, 2, 3})


fst.add(4)  # Error (we cannot modify frozenset)
print(fst)
```

# Use Cases of Sets

```python
num = [1, 2, 2, 3, 4, 4, 5]
print("num",num)

unique = list(set(num))  # [1,2,3,4,5]
print("unique",unique)
```

```python
if 3 in {1,2,3,4,5}:
    print("Found")
```

```python
text = "python  python is great"
print(set(text.split()))  # {'great','is','python'}
```
