# Built-in Exceptions

```python
| Exception           | Description                              |
| ------------------- | ---------------------------------------- |

|  ZeroDivisionError  | Raised when dividing by zero             |

|  ValueError         | Raised when a function gets invalid data |

|  TypeError          | Raised when a wrong type is used         |

|  IndexError         | List index out of range                  |

|  KeyError           | Dictionary key not found                 |

|  FileNotFoundError  | Raised when a file can’t be found        |

|  ImportError        | Module import fails                      |

|  AttributeError     | Invalid attribute reference              |
```

# Syntax :

```python
try:
    # Code that might raise an exception
    risky_code()

except SomeException as e:
    # Code to handle the exception
    print(f"An error occurred: {e}")

else:
    # Runs if no exception occurs
    print("No errors!")

finally:
    # Always runs, even if there's an exception
    print("This will run no matter what.")
```
